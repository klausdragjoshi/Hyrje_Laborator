*   **Vizualizimi:**   Përdorimi i *Matplotlib* për paraqitjen grafike të të dhënave në mënyrë shkencore (boshtet, titujt, legjendat).

**Ushtime me Stringje,Funksione dhe librarite Numpy/Matplotlib**

```python
materialet = [("Alumin", 2700), ("Hequr", 7874.234), ("Bakër", 8960.5)]

print(f"{'Materiali':<15} | {'Dendësia (kg/m³)':>15}")
print("-" * 35)

for emri, dendesia in materialet:
    print(f"{emri:<15} | {dendesia:>15.2f}")
```

```python
import math

def kontroll_vlerash(vlefta):
    if vlefta > 0:
        return f"Rrënja: {math.sqrt(vlefta):.2f}"
    elif vlefta < 0:
        return f"Katrori: {vlefta**2}"
    else:
        return "Vlera është zero"

test_numrat = [9, -4, 0, 25, -3]
for n in test_numrat:
    print(f"Numri {n} -> {kontroll_vlerash(n)}")
```

```python
import numpy as np

# Krijimi i array-t
x = np.linspace(0, 2 * np.pi, 20)

# Llogaritjet
y_sin = np.sin(x)
y_cos = np.cos(x)

print(f"Sinusi Max: {np.max(y_sin):.2f}")
print(f"Sinusi Min: {np.min(y_sin):.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Ndërtimi i grafikëve
plt.plot(x, y_sin, 'b--', label='Sinusi (x)')
plt.scatter(x, y_cos, color='red', label='Kosini (x)')

# Stilimi
plt.title("Analiza e Funksioneve Trigonometrike", fontsize=14)
plt.xlabel("Vlera e x (radian)", fontsize=12)
plt.ylabel("Amplituda", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

# Shfaqja
plt.show()
```

```python
# Të dhënat hyrëse
rryma = 0.005234  # Amper (A)
rezistenca = 1200 # Ohm (Ω)

# Llogaritja
tensioni = rryma * rezistenca

# Formatimi me Print f-string
print(f"{' ANALIZA E QARKUT ':=^30}")

# Rreshtimi i tekstit majtas dhe numrave djathtas
print(f"{'Rryma:':<15} {rryma:>10.5f} A")
print(f"{'Rezistenca:':<15} {rezistenca:>10.1f} Ω")

print("-" * 30)

# Rezultati final me ngjyrë/theksim imagjinar (duke përdorur simbole)
print(f"{'TENSIONI:':<15} {tensioni:>10.3f} V")
print(f"{'='*30}")
```

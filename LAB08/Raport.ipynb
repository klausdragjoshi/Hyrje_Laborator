### 1. Qëllimi i Laboratorit
Në këtë detyrë synojmë:
1. Matjen e drejtpërdrejtë të masës ($m$) dhe rrezes ($R$) së sferës së lavjerrësit.
2. Llogaritjen e dendësisë së materialit përmes formulës $\rho = m/V$.
3. Analizimin e përhapjes së pasigurisë (gabimit) nga madhësitë bazë te madhësia e derivuar (dendësia).

### 2. Baza Teorike dhe Formulat
Dendësia e sferës llogaritet si:
$$\rho = \frac{m}{\frac{4}{3}\pi R^3}$$

Për vlerësimin e pasigurisë, përdorim metodën e përhapjes së pasigurisë relative:
$$\frac{\Delta \rho}{\rho} = \sqrt{\left( \frac{\Delta m}{m} \right)^2 + \left( 3 \frac{\Delta R}{R} \right)^2}$$

Ku:
* $\Delta m$ është pasiguria e peshores.
* $\Delta R$ është pasiguria e mikrometrit ose kalibrit.

### 3. Të Dhënat Eksperimentale
Masa e sferës u mat $m = 150.25$ g me pasiguri $\Delta m = 0.01$ g.

| Nr. | Rrezja $R$ (mm) | Pasiguria $\Delta R$ (mm) |

*Vlera mesatare e rrezes:* $R_{mes} = 15.016$ mm

### 4. Përpunimi i Rezultateve
Duke përdorur vlerat mesatare:
* $m = 0.15025$ kg
* $R = 0.015016$ m

*Vëllimi:* $V = \frac{4}{3}\pi R^3 \approx 1.419 \times 10^{-5} \text{ m}^3$  
*Dendësia:* $\rho = \frac{m}{V} \approx 10588.4 \text{ kg/m}^3$

*Pasiguria relative:*
$\epsilon_{\rho} = \sqrt{(\frac{0.00001}{0.15025})^2 + (3 \frac{0.00001}{0.015016})^2} \approx 0.002$ (ose 0.2%)

*Pasiguria absolute:*
$\Delta \rho = 10588.4 \times 0.002 \approx 21.2 \text{ kg/m}^3$

### 5. Konkluzione
Pas analizës së të dhënave, dendësia e sferës së lavjerrësit rezultoi:
*$$\rho = (10588.4 \pm 21.2) \text{ kg/m}^3$$*

Vërejmë se gabimi në matjen e rrezes ka peshën kryesore në pasigurinë totale, pasi rrezja në formulën e dendësisë është në fuqi të tretë. Rezultati është konsistent me materialin e plumbit.

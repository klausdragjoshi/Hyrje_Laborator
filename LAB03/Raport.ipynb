### Analiza Matematike
Për të llogaritur zhvendosjen në lëvizjen drejtvizore të njëtrajtshme të nxituar, përdorim ekuacionin:

$$s = v_0 t + \frac{1}{2} a t^2$$

Ku:
* $s$ është zhvendosja.
* $v_0$ është shpejtësia fillestare.
* $a$ është nxitimi (për rënie të lirë përdorim $g \approx 9.81 \, m/s^2$).

Gjithashtu, mund të shprehim energjinë e sistemit përmes integralit:
$$W = \int_{a}^{b} F(x) \, dx$$

```python
# Definimi i konstanteve
g = 9.81  # m/s^2
t = 5     # sekonda
v0 = 0    # shpejtesia fillestare

# Llogaritja e rrugës së kaluar gjatë rënies së lirë
lartesia = v0 * t + 0.5 * g * (t**2)

print(f"Pas {t} sekondash, objekti ka rënë nga një lartësi prej {lartesia:.2f} metrash.")
```

### Llogaritja e Periodës së Lavjerrësit
Formula për periodën e një lavjerrësi matematik është:

$$T = 2\pi \sqrt{\frac{L}{g}}$$

Ku:
* $L$ është gjatësia e fijeve (metra).
* $g$ është nxitimi i rënies së lirë ($\approx 9.81 \, m/s^2$).

```python
# Ushtrimi: Llogaritja e forcës (Ligji i dytë i Newton-it)
massa = float(input("Vendosni masën e objektit (kg): "))
nxitimi = float(input("Vendosni nxitimin (m/s^2): "))

forca = massa * nxitimi

print("-" * 30)
print(f"Për masën {massa} kg dhe nxitimin {nxitimi} m/s^2:")
print(f"Forca e ushtruar është F = {forca} Newton")
print("-" * 30)
```

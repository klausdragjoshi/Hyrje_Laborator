*Data:* 2026-03-23 


### 1. Objektivi
Qëllimi i këtij laboratori është njohja me sistemin e kontrollit të versioneve (Git) 
dhe platformën GitHub për bashkëpunim dhe ruajtje të projekteve.

2. Teoria e Sistemit Git 

* Git: Programi që mban mend ndryshimet në kompjuter (lokalisht).
* GitHub: Faqja ku ruhet kodi në internet (remote).
* Repository: Dosja e projektit që ndiqet nga Git.
* Commit: Një "foto" e gjendjes së projektit në një moment të caktuar.

3. Logu i Punës 

1. Krijimi i repository-t në GitHub me emrin Hyrje_Laborator.
2. Konfigurimi i .gitignore për të shmangur file-t e panevojshëm.
3. Krijimi i strukturës së dosjeve me nën-dosjet data/ dhe figs/.

### Tabelë: Komandat Kryesore të Git

| Komanda | Përshkrimi i Funksionit |
| :--- | :--- |
| git init | Inicializon një repository të ri lokal. |
| git clone <url> | Kopjon një repository ekzistues nga GitHub në kompjuter. |
| git add . | Shton të gjitha ndryshimet në "Staging Area". |
| git commit -m "mesazhi" | Ruante ndryshimet me një përshkrim specifik. |
| git push origin main | Dërgon ndryshimet e bëra në serverin e GitHub. |
| git pull | Merr ndryshimet e fundit nga GitHub në kompjuter. |
| git status | Shfaq gjendjen aktuale të file-ve (ashtu siç janë modifikuar). |

### Përfundime
Përdorimi i Git është esencial për çdo programues. Përmes këtij laboratori, 
mësova se si të mbaj një projekt të organizuar dhe si të dokumentoj çdo 
ndryshim përmes "commits". Respektimi i strukturës së dosjeve ndihmon në 
automatizimin e kontrollit të punës sime.
